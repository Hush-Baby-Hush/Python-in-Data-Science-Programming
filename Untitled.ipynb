```python
def ternary_to_decimal(x):
    # enter your code below this line
    ans = 0
    temp = 0
    while (x):
        a = x%10
        ans+=a*3**temp
        temp+=1
        x = (x-a)/10
    return int(ans)
```

```python
def first_three_prime(nums):
    # enter your code below this line
    array=[]
    for i in nums:
        if len(array)==3:
            return array
        if i<=1:
            continue
        if i==2:
            array.append(i)
        flag = 1
        for j in range(2,i):
            if i%j==0:
                flag=0
                break
        if flag==1:
            array.append(i)
    return array
```

```python
def datecheck(x):
    if pd.isna(x[0]):
        return np.nan
    if x[0][0]=='0':
        x[0] = x[0][1]
    if x[1][0]=='0':
        x[1] = x[1][1]
    try:
        datetime.datetime(year=int(x[2]),month=int(x[0]) ,day=int(x[1]))
    except:
        return np.nan
    if len(x[0])==1:
        x[0] = '0'+x[0]
    if len(x[1])==1:
        x[1] = '0'+x[1]      
    return x[0]+'/'+x[1]+'/'+x[2]

```
