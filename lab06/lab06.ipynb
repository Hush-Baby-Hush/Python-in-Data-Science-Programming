## 1. DataFrame Filtering

In this lab, we will again be working with the cleaned version of the Ramen Ratings dataset. There are 5 variables:

  - `'Brand'`: the brand of the ramen product.
  - `'Variety'`: the product’s name on its label.
  - `'Style'`: packaging style (cup, bowl, tray, etc.)
  - `'Country'`: country or dependent territory the product was produced in.
  - `'Stars'`: the rating of the product on a 5-point scale.

This dataset is available at: https://stat430.hknguyen.org/files/datasets/clean-ramen.csv


Load the Ramen Ratings dataset to Python and store it in a variable called `ramen`.

```python
# your code here
import pandas as pd
ramen = pd.read_csv('https://stat430.hknguyen.org/files/datasets/clean-ramen.csv')
ramen
```

### Exercise 1.1 (0.5 point)

Filter only the observations where the ramen is made in Japan and in Cup, OR the ramen is made in Thailand and in Pack.

```python
# your code here
result = ramen[((ramen['Style'] == 'Cup') & (ramen['Country'] == 'Japan')) | ((ramen['Style'] == 'Pack') & (ramen['Country'] == 'Thailand'))]
result
```

### Exercise 1.2 (0.5 point)

Filter only the observations where the ramen had **at least 4 stars** and it is made in `USA` or `United States`.

```python
# your code here
result = ramen[ramen.Country.isin(['USA', 'United States']) & (ramen['Stars'] >= 4)]
result
```

### Exercise 1.3 (0.5 point)

Filter only the `Brand` and `Variety` attributes (columns) of the observations where the ramen had at least 4 stars and it is made in `Vietnam`.

```python
# your code here
result = ramen[(ramen['Stars'] >= 4)  & (ramen['Country'] == 'Vietnam')][['Brand','Variety']]
result
```

### Exercise 1.4 (0.5 point)

Filter only the `Variety` and `Stars` attributes of the observations where the ramen had **at most 2 stars** and the brand is `Nissin`.

```python
# your code here
result = ramen[(ramen['Stars'] <= 2)  & (ramen['Brand'] == 'Nissin')][['Variety','Stars']]
result
```

## 2. Computing Basic Statistics on Subset of a DataFrame

### Exercise 2.1 (0.5 point)

Compute the mean stars rating of all ramen made in the US (that is `USA` and `United States`).

```python
# your code here
result = ramen[ramen.Country.isin(['USA', 'United States'])]['Stars'].mean()
result
```

### Exercise 2.2 (0.5 point)

Find the maximum stars rating of a Vifon ramen (of brand Vifon).

```python
# your code here
result = ramen[ramen['Brand'] == 'Vifon'].sort_values(by = 'Stars', ascending=False).head(1)
result[['Stars']]
```

### Exercise 2.3 (0.5 point)

Find the observation (meaning the whole row) of the a Nongshim ramen with the lowest stars rating.

```python
# your code here
result = ramen[ramen['Brand'] == 'Nongshim'].sort_values(by = 'Stars').head(1)
result
```

### Exercise 2.4 (0.5 point)

Now, find the observation corresponding to the value found in Ex 3.3. That is, find and print the whole row of the Vifon ramen that has stars rating of 5.0.

```python
# your code here
result = ramen[(ramen['Brand'] == 'Vifon') & (ramen['Stars'] == 5)]
result
```

## 3. Function Application

### Exercise 3.1 (2 points)

Compute the average rating of all **beef-flavored** ramen in the US. That is,

- Find all the ramen observation produced in the US (only 'USA') with the word **beef** (`beef`, `Beef`, or `BEEF` must all be counted) in its label.
- Compute the average star rating of those observations.

**Notes: The use of loops is not allowed in this exercise. Vectorized string functions such as `str.func_name()` are not allowed either (since it has not been covered in this course, we will learn it in future weeks).**

```python
# your code here
f = lambda x:  'beef' in x['Variety'].lower()
result = ramen[(ramen['Country'] == 'USA') & ramen.apply(f, axis = 1)]
result[['Stars']].mean()
```

### Exercise 3.2 (2 points)

Ha does not like chicken-flavored ramen as much as other types. So, she wants you to adjust the ratings of **chicken-flavored** ramen observations. That is, the ramen label contains the word **chicken** (`chicken`, `Chicken`, or `CHICKEN` must all be counted). If the rating of a chicken-flavored ramen is less than 1, change it to 0. If it is at least 1, lower the rating by 1.

**Notes: The use of loops is not allowed in this exercise. Vectorized string functions such as `str.func_name()` are not allowed either (since it has not been covered in this course, we will learn it in future weeks).**

```python
# your code here
def f(x):
    if 'chicken' in x['Variety'].lower():
        x['Stars'] = max(0,x['Stars']-1)
    return x
result = ramen.apply(f, axis = 1)
result
```
