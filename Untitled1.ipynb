```python
import pandas as pd

cleveland = pd.read_csv('processed.cleveland.data', header=None,
                       names=['age', 'sex', 'cp', 'trestbps', 'chol', 'fbs', 'restecg',
                              'thalach', 'exang', 'oldpeak', 'slope', 'ca', 'thal', 'num'],
                       na_values=['?'])
def f(x):
    if x['slope']==1.0 : #upsloping
        if x['thalach'] >=160.0:
            x['thalach_level'] = 'high'
        else:
            x['thalach_level'] = 'low'
    if x['slope']==2.0 : #flat
        if x['thalach'] >=140.0:
            x['thalach_level'] = 'high'
        else:
            x['thalach_level'] = 'low'
        
    if x['slope']==3.0 : #downsloping
        if x['thalach'] >=145.0:
            x['thalach_level'] = 'high'
        else:
            x['thalach_level'] = 'low'
    return x

cleveland = cleveland.apply(f, axis = 1)
cleveland['num'] = cleveland['num'].astype('int64')
```

```python
import numpy as np
import pandas as pd
import re
phone_book = pd.read_csv('phone-book.csv')
phone_book['phone'] = phone_book['phone'].str.strip()
phone_book
```

```python
def join_phone(x):
    result = x[0]+x[1]+x[2]
    if len(result)!=10:
        return np.nan
    if result[0:3]<'201' or result[0:3]>'989':
        return np.nan
    return result[0:3]+'-'+result[3:6]+'-'+result[6:]
pattern = '[-\s()]*([0-9]+)[-\s()]*([0-9]+)[-\s()]*([0-9]+)'
data = phone_book['phone'].str.extract(pattern, flags=re.IGNORECASE)
data
```

```python
phone_book['phone'] = data.agg(join_phone, axis=1)
phone_book
```
